```python
# Import libraries

import numpy as np
import pandas as pd
import csv
```

```python
# open csv file

data = open('DiaphragmUltrasoundI_DATA_2020-11-30_2048.csv', encoding='utf-8')
csv_data = csv.reader(data)
```

```python
# reformat into python object list of lists
data_lines = list(csv_data)
```

```python
# row 0 is the variable names 
# print out the index of each variable

print('dataset variables: ')
n=0
for lines in data_lines[0]:
    print(str(n) + ': ' + str(data_lines[0][n]))
    n+=1

# make variable dictionary: variable name = key, and index = value

var_dict = {}
k=0
for variable in data_lines[0]:
    var_dict[data_lines[0][k]]=k
    k+=1

# height_cm was misspelled as hieght_cm, need to use the wrong spelling as key in var_dict
```

```python
# dataset entries(rows): subjects# + Udi#
len(data_lines)
```

```python
# print the 1st entry
for line in data_lines[:2]:
    print(line)
```

```python
# variables in clean_dataset_1

clean_dataset_1_variables = [
'subject_id', 
'redcap_repeat_instrument', 
'redcap_repeat_instance',
'age_years',
'sex',
'hieght_cm',
'weight',
'charlson_comorbidity_index',
'us_serial_num_udi',
'date_udi_assessment',
'thickness_ex_tidal',
'thickness_in_tidal',
'thickness_fraction_tidal',
'excursion_tidal',
'contraction_vmax',
'relaxation_vmax',
'acceleration_time',
'relaxation_rate',
'contraction_vti',
'relaxation_vti',
'vent_mode_udi',
'fio2_udi',
'o2_sat_pulsox_udi',
'paralytics_udi',
'hos_admission_date',
'number_of_icu_stay',
'day_mortality',
'discharge_disposition',
'date_of_death',
'transfer_to_icu_date',
'time_of_intubation_icu',
'time_of_extubation_icu',
'respiratory_status_at_24_h',
'respiratory_status_at_48_h',
'respiratory_status_at_72_h',
'transfer_out_of_icu_date'
]

print(clean_dataset_1_variables)

# get the variable index for variables in clean_dataset_1
var_index = []
for variable in clean_dataset_1_variables:
    var_index.append(var_dict[variable])
print(var_index)

clean_dataset_1 = []
clean_dataset_1.append(clean_dataset_1_variables)

for row in range(len(data_lines)):
    subject_data = []
    for index in var_index:
        subject_data.append(data_lines[row][index])
    if subject_data[1]== '' or subject_data[1]== 'ultrasound_measurements':
        clean_dataset_1.append(subject_data)

print(len(clean_dataset_1))
```

```python
# fill the blanks of ultrasound instrument
# i.e. combine patient demographic, comorbidity, and outcome data with ultrasound instrument

subject_char = []
subject_char_index = []

for row in range(1,len(clean_dataset_1)):
    if clean_dataset_1[row][1]=='': # ie: not an instrument entry
        subject_char = []
        subject_char_index = []
        subject_char = clean_dataset_1[row][:]
        for column in subject_char:
            subject_char_index.append(column != '')
    elif clean_dataset_1[row][1]=='ultrasound_measurements': # an Udi instrument entry
        udi_instrument = []
        udi_instrument = clean_dataset_1[row][:]
        if udi_instrument[0]==subject_char[0]: # instrument subject_id = subject_char subject_id
            for n in range(1, len(subject_char_index)): # fill subject characteristics into the blanks in udi_instrument
                if subject_char_index[n]:
                    clean_dataset_1[row][n]=subject_char[n]
                #finished updating the clean_dataset_1
                #print(clean_dataset_1[row])

print(clean_dataset_1[2])
print(len(clean_dataset_1))
print(clean_dataset_1[0])
```

```python
# derived variables
# 1. indexing the dataset
dataset_dict = {}
k=0
for variable in clean_dataset_1[0]:
    dataset_dict[clean_dataset_1[0][k]]=k
    k+=1
print(dataset_dict)

# intubation/vent day
# get column indext for variables
time_of_intubation_icu_index = dataset_dict['time_of_intubation_icu']

date_udi_assessment_index = dataset_dict['date_udi_assessment']


# create new column and hosting derived variable
clean_dataset_1[0].append('vent_day')
dataset_dict['vent_day']=len(clean_dataset_1[0])
print(dataset_dict['vent_day'])

from datetime import date

for row in range(1, len(clean_dataset_1)):
    if clean_dataset_1[row][time_of_intubation_icu_index]!='' and clean_dataset_1[row][date_udi_assessment_index]!='':
        intubation_date = date(int(clean_dataset_1[row][time_of_intubation_icu_index][0:4]), int(clean_dataset_1[row][time_of_intubation_icu_index][5:7]), int(clean_dataset_1[row][time_of_intubation_icu_index][8:10]))
        udi_date = date(int(clean_dataset_1[row][date_udi_assessment_index][0:4]), int(clean_dataset_1[row][date_udi_assessment_index][5:7]), int(clean_dataset_1[row][date_udi_assessment_index][8:10]))
        vent_day_udi = (udi_date-intubation_date).days
        clean_dataset_1[row].append(str(vent_day_udi))
    else:
        clean_dataset_1[row].append('')
    #print(clean_dataset_1[row])

print(clean_dataset_1[0])

# hospital/icu @ Udi measurement...pending
```
